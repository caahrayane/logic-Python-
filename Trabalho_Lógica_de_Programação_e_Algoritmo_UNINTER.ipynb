```python
#Código 1: Mensagem de boas-vindas com o nome
print("Bem-vindo a nossa loja de delivery Açaí e Cupuaçu! - Desenvolvido por Camilly Rayane")
print("--------------------CARDÁPIO--------------------")
print("-----| TAMANHO | CUPUAÇU (CP) | AÇAÍ (AC) |-----")
print("-----|    P    |   R$ 10,00   | R$ 12,00  |-----")
print("-----|    M    |   R$ 15,00   | R$ 17,00  |-----")
print("-----|    G    |   R$ 19,00   | R$ 21,00  |-----")
print("------------------------------------------------")

#Código: Somar os valores dos produtos
total_pedido = 0

#Código 2: sabor e validação
while True:
    sabor = input("Digite o sabor desejado (CP para Cupuaçu, AC para Açaí): ")
    if sabor not in ['CP', 'AC']:
        print("Sabor inválido. Tente novamente.")
        continue
#Código 3: Input tamanho e validação
    print("")
    tamanho = input("Digite o tamanho desejado (P, M ou G): ")
    tamanho = tamanho.upper()
    if tamanho not in ['P', 'M', 'G']:
        print("Tamanho inválido. Tente novamente.")
        continue

#Código 4: Combinação de tamanhos e sabores
    if sabor == 'CP':
        if tamanho == 'P':
            valor_pedido = 10
        elif tamanho == 'M':
            valor_pedido = 15
        elif tamanho == 'G':
            valor_pedido = 19
    elif sabor == 'AC':
        if tamanho == 'P':
            valor_pedido = 12
        elif tamanho == 'M':
            valor_pedido = 17
        elif tamanho == 'G':
            valor_pedido = 21
#Código: Somar os valores dos produtos
    total_pedido += valor_pedido

#Código 7: Pedido final
    if total_pedido > 0:
        print(f"Você pediu {'Cupuaçu' if sabor == 'CP' else 'Açaí'} de tamanho {tamanho} e o valor é R${valor_pedido:.2f}")

#Código: Input para inserir algo a mais no pedido
    mais_pedidos = input("Deseja pedir mais alguma coisa? (Digite 'sim' ou 'nao'): ")
    if mais_pedidos.lower() != 'sim':
      print("")
      break

#Código 6: Total do pedido
print(f"Valor total do pedido: R${total_pedido:.2f}")
```

#3º Exercício da Atividade Prática

```python
#Código 1: Mensagem de boas-vindas com o nome
print("Bem-vindo ao sistema de cobrança da copiadora - Desenvolvido por Camilly Rayane")
print("")

#Código 2: Função escolha o serviço
def escolha_servico():
    while True:
        print("Escolha o serviço:")
        print("DIG - Digitalização")
        print("ICO - Impressão Colorida")
        print("IBO - Impressão Preto e Branco")
        print("FOT - Fotocópia")

        servico = input("Digite o código do serviço desejado: ").upper()

        if servico in ['DIG', 'ICO', 'IBO', 'FOT']:
            return servico
        else:
            print("Opção inválida. Tente novamente.")
            print("")

#Código 3: Função número de páginas
def numero_pagina():
    while True:
        try:
            numero_paginas = int(input("Digite o número de páginas: "))
            print("")
            if numero_paginas < 10:
                return numero_paginas
            elif 10 <= numero_paginas < 100:
                return int(numero_paginas * 0.9)  # desconto de 10%
            elif 100 <= numero_paginas < 1000:
                return int(numero_paginas * 0.85)  # desconto de 15%
            elif 1000 <= numero_paginas < 10000:
                return int(numero_paginas * 0.8)  # desconto de 20%
            else:
                print("Número de páginas não aceito. Tente novamente.")
        except ValueError:
            print("Por favor, insira um valor numérico válido.")
            print("")

#Código 4: Função serviço extra
def servico_extra():
    while True:
        print("Escolha o serviço adicional:")
        print("1 - Encadernação Simples R$ 10,00")
        print("2 - Encadernação de Capa Dura R$ 25,00")
        print("0 - Para Nenhum")

        adicional = input("Digite o código do serviço adicional desejado: ")

        if adicional in ['1', '2', '0']:
            return adicional
        else:
            print("Opção inválida. Tente novamente.")
            print("")

#Código 5: Obter valores do usuário
servico = escolha_servico()
numero_paginas = numero_pagina()
adicional = servico_extra()

#Código 6: Cálculo do total a pagar
try:
    if servico == 'DIG':
        valor_pagina = 1.10
    elif servico == 'ICO':
        valor_pagina = 1.00
    elif servico == 'IBO':
        valor_pagina = 0.40
    elif servico == 'FOT':
        valor_pagina = 0.20

    total_pagar = valor_pagina * numero_paginas

#Código 7: Adicional ao total
    if adicional == '1':
        total_pagar += 10
    elif adicional == '2':
        total_pagar += 25

#Código 8: Impressão do total a pagar
    print(f"Total a pagar: R${total_pagar:.2f}")
    print("")

except Exception as e:
    print(f"Erro: {e}")


#Código 9: Saída de console
print("Obrigado por utilizar nossos serviços!")
print("")
```

#4º Exercício da Atividade Prática


```python
# Código 1: Mensagem de boas-vindas com o nome
print("Bem-vindo ao sistema de gerenciamento de livros - Desenvolvido por Camilly Rayane")
print("")

# Código 2: Variáveis globais
lista_livro = []
id_global = 0

# Código 3: Função cadastrar livro
def cadastrar_livro(id):
    global id_global
    id_global += 1
    nome = input("Digite o nome do livro: ")
    autor = input("Digite o autor do livro: ")
    editora = input("Digite a editora do livro: ")
    print("")

    livro = {
        "id": id,
        "nome": nome,
        "autor": autor,
        "editora": editora
    }

    lista_livro.append(livro)
    print(f"Livro cadastrado com ID {id}")

# Código 4: Função consultar livro
def consultar_livro():
    while True:
        print("Opções de consulta:")
        print("1. Consultar Todos")
        print("2. Consultar por Id")
        print("3. Consultar por Autor")
        print("4. Retornar ao menu")

        opcao = input("Digite a opção desejada: ")

        if opcao == '1':
            print("Consulta de Todos os Livros:")
            for livro in lista_livro:
                print(livro)
        elif opcao == '2':
            id_consulta = int(input("Digite o Id do livro: "))
            livro_encontrado = False
            for livro in lista_livro:
                if livro['id'] == id_consulta:
                    print(livro)
                    livro_encontrado = True
                    break

            if not livro_encontrado:
                print("Livro não encontrado.")
        elif opcao == '3':
            autor_consulta = input("Digite o autor do livro: ")
            for livro in lista_livro:
                if livro['autor'] == autor_consulta:
                    print(livro)
            else:
                print("Nenhum livro encontrado para este autor.")
        elif opcao == '4':
            break
        else:
            print("Opção inválida. Tente novamente.")

# Código 5: Função remover livro
def remover_livro():
    id_remover = int(input("Digite o ID do livro a ser removido:"))
    for i, livro in enumerate(lista_livro):
        if livro['id'] == id_remover:
            del lista_livro[i]
            print(f"Livro com ID {id_remover} removido com sucesso.")
            break
    else:
        print(f"Livro com ID {id_remover} não encontrado.")

# Código 6: Main - Estrutura de Menu
while True:
    print("--------------------------------- MENU PRINCIPAL --------------------------------")
    print("\nOpções do Menu:")
    print("1. Cadastrar Livro")
    print("2. Consultar Livro")
    print("3. Remover Livro")
    print("4. Encerrar Programa")

    opcao_menu = input("Digite a opção desejada: ")
    print("")
    print("------------------------------- MENU CADASTRO LIVRO -----------------------------")
    print("")

    if opcao_menu == '1':
        cadastrar_livro(id_global)
    elif opcao_menu == '2':
        consultar_livro()
    elif opcao_menu == '3':
        remover_livro()
    elif opcao_menu == '4':
        print("Programa encerrado. Obrigado por utilizar nosso sistema.")
        break
    else:
        print("Opção inválida. Tente novamente.")
```
